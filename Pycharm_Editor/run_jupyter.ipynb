```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

url = 'https://people.sc.fsu.edu/~jburkardt/data/csv/airtravel.csv'
data = pd.read_csv(url)

print("Dataset Information:")
print(data.info())
print("\nDescriptive Statistics:")
print(data.describe())

print("\nMissing Values:")
print(data.isnull().sum())

# Convert year columns to int
data.columns = ['Month', '1958', '1959', '1960']
data['1958'] = data['1958'].astype(int)
data['1959'] = data['1959'].astype(int)
data['1960'] = data['1960'].astype(int)

# Plotting the data
plt.figure(figsize=(14, 7))
sns.lineplot(data=data.set_index('Month'))
plt.title('Air Travel Passengers Over Years')
plt.ylabel('Number of Passengers')
plt.show()

sns.pairplot(data.set_index('Month').reset_index())
plt.show()
```
